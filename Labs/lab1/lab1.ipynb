##### NumPy (створити набір даних довільного розміру, провести CRUD операції над даними (скаляр, вектор, матриця)). Використати arange, random.

```python
import numpy as np

dataset = np.arange(7)
datasetR = np.random.randint(0, 20, 7)
print(f"Dataset 1 (arange): {dataset}")
print(f"Dataset 2 (random): {datasetR}\n")

scalar = np.array(25)
print(f"Initial scalar: {scalar}")
scalar = np.dot(dataset, datasetR)
print(f"Updated scalar: {scalar}\n")

vector = np.arange(0, 50, 10)
print(f"Initial vector: {vector}")
print(f"Subvector: {vector[1:4]}")
vector[3] = 75
print(f"Updated vector: {vector}")
vector = np.delete(vector, [0, 2])
print(f"Vector with 1st and 3rd elements removed: {vector}\n")

matrix = np.random.randint(0, 13, (5, 5))
print(f"Matrix:\n {matrix}")
print(f"\nSubmatrix:\n {matrix[1:4, 1:4]}")
print(f"\nSubmatrix(2nd row):\n {matrix[1, :]}")
print(f"\nSubmatrix(3rd column):\n {matrix[:, 2]}")
matrix[2, 1] = 55 #Changing element in the 3rd row, 2nd column
print(f"\nUpdated matrix:\n {matrix}")
matrix = np.delete(matrix, 4, 0)
print(f"\nMatrix with 5th row removed:\n {matrix}")
matrix = np.delete(matrix, 0, 1)
print(f"\nMatrix with 1st column removed:\n {matrix}")

```

##### Matplotlib (створити набір даних лінійної функції з урахуванням помилок вимірювань, побудувати графік функції та згенерованих даних, обчислити похибку з використанням метрик MAE, MSE). Записати отримані результати у csv файл (формат: X,Y, Y_hat, mAE, mSE)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def linear_function(x):
    return 3 * x - 4

np.random.seed(0)
x = np.linspace(0, 10, 100)
y_hat = linear_function(x) # predicted values of y
noise = np.random.normal(0, 1, 100)
y = linear_function(x) + noise # values of y affected by "noise"

mae = np.mean(np.abs(y - y_hat))
mse = np.mean((y - y_hat) ** 2)
print(f"MAE (Mean Absolute Error): {mae}")
print(f"MSE (Mean Squared Error): {mse}")

# Making a plot
plt.figure(figsize=(10, 6))
plt.plot(x, y_hat, label='Linear function', color='blue', linewidth=3)
plt.scatter(x, y, label='Function with measurement errors', color='red', alpha=0.8)
plt.title('Plot of a linear function and a function with measurement errors')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Writing the results to a CSV file
data = {'X': x, 'Y': y, 'Y_hat': y_hat, 'MAE': mae, 'MSE': mse}
df = pd.DataFrame(data)
df.to_csv('results.csv', index=False)
```
